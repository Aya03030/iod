```python
ax = df['quality'].value_counts().sort_index().plot(kind='bar', title='Wine Quality')

ax.set_xlabel('Quality')
ax.set_ylabel('Counts')
```

```python
df.corr()
```

```python
plt.figure(figsize=(10,10))
heatmap = sns.heatmap(df.corr(), annot=True)
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=50);
```

```python
df.groupby('quality')['alcohol'].mean()
```

```python
# 
alcohol_means = df.groupby('quality')['alcohol'].mean()

# 
plt.bar(alcohol_means.index, alcohol_means.values)

# 
plt.xlabel("Quality")
plt.ylabel("Average Alcohol")

# 
plt.grid(True)

# 
plt.show()
```
